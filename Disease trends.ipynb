```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Hospital.csv')
df
```

```python
df.info()
```

```python
df.drop(columns = ['follow_up_required','recovery_days','cause_of_death','death_flag','outcome','income_level','bmi','comorbidity','symptoms','severity','region','age_group','alcohol_use'],axis = 1 , inplace = True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# changing dataype 
df['Diagnosis_Date'] = pd.to_datetime(df['Diagnosis_Date'])
```

```python
# creating Age_Group 
df['age_group'] = np.where(df['age'] <= 12 ,'Child',np.where(df['age'] <= 19 , 'Teen',np.where(df['age'] <= 59 ,'Adult','Senior')))
```

```python
df[['age','age_group']]
```

```python
df.columns = df.columns.str.strip().str.title()
```

```python
#Fixing Text column
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip().str.title()
```

```python
df
```

```python
#Validate Numeric Columns
df = df[df['Age'] > 0]
df = df[df['Treatment_Cost_Inr'] > 0]
```

```python
#Final look Datatypes and Data 
df.info()
```
